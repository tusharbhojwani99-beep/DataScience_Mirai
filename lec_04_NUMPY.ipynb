```python
# WHERE ()

a = np.random.randint(2,15,6)
b = np.random.randint(1,19,6)

c = np.where(b>2, b, a)
print(c)
```

```python
a = np.array([[3, 5, 11],
              [9, 15, 7]])

b = np.array([[1, 35, 11],
              [9, 75, 91]])

print(np.sort(a, axis=0))
print(" - ")
c = np.where(b>2, b, a)
print(c)
```

```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

c = np.matmul(a, b)          # matmul() --------> used for matrix multiplication

print(c)
```

```python
#   {  "@" ------------> also used instead of matmul() }

a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

print(a @ b)
```
