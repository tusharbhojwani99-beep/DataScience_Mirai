```python
lst = [1,2,3,'hello','c',[2,4]]
print(lst)

# append {add element or another list at last of the previous list}

lst.append([10,12,14])
print(lst)

# extend {add the other whole list in last as separate element}

lst.extend([4,5,6])
print(lst)
```

```python
# Tuppels

tp = (1,20,33,4,59)
a = sorted(tp)             #creates a new list which will be sorted in nture
print(tp)
print(a)
```

```python
tp = (1,20,20,33,4,1,59,3,3,3)
a = tuple(sorted(tp))
print(tp,"     (sorted) ---->     ",a)
```

```python
# FUNCTIONS USED IN TUPLES

# Category                  Function

# count values              count()
# find                      index()
# size                      len()
# total                     sum()
# highest/lowest            max()/min()
# sort                      sorted(){returns list}
```

```python
# count()
tp = (1,20,20,33,4,1,59,3,3,3)
a = tp.count(20)
print(a)
```

```python
# index()
tp = (1,20,20,33,4,1,59,3,3,3)
a = tp.index(33)
print(a)
```

```python
# len
tp = (1,20,20,33,4,1,59,3,3,3)
a = len(tp)
print(a)
```

```python
# max()/min()
tp = (0,1,20,20,33,4,1,59,3,3,3)
a = max(tp)
b = min(tp)
print("maximum = ",a)
print("minimum = ",b)
```

```python
n = int(input("Enter size of lst 1: "))
m = int(input("Enter size of lst 2: "))

lst1 = []
lst2 = []

# lst1 ke liye inputs lo
print("Enter elements for list 1:")
for i in range(n):
  lst1.append(int(input()))

# lst2 ke liye inputs lo
print("Enter elements for list 2:")
for i in range(m):
  lst2.append(int(input()))

lstF = []

# Dono ka size check karo
if n != m:
  print("not valid")
else:
  for i in range(n):
    lstF.append(lst1[i] + lst2[i])

print("Final List:", lstF)
```
