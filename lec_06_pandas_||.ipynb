```python
# Function	 Meaning

# isnull() ->  missing values check
# notnull()->  not missing values
# unique() ->  all distinct values
# nunique()->  count of unique values
```

```python
data = {
    "name": ["Aman", "Riya", "Vikram", None],
    "marks": [85, np.nan, 78, 90]
}

df = pd.DataFrame(data)

print(df.isnull())
df['marks'].nunique()
# print(df['marks'].unique())

df.isnull().sum()               #----> Count missing values
df[df['marks'].notnull()]               #-> Keep only non-null rows
```
