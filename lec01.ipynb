```python
lst = [ 1, 2, 3, 'hi', 'c', [ 2 , 4 ] ]

print(lst)

#append() ---> use to add the list at the end ( list--> [a,b,c])

lst.append([2 , 3, 4 ])
print(lst)

#extend() ---> use to add the elemnts at the end of the block ( diffrent from append )

lst.extend([2 , 3, 4])
print(lst)

#insert()

#lst.insert(index , value)

lst.insert(2 , "value")
print(lst)


#pop()

lst.pop()
print(lst)


#remove()

lst.remove(2)
print(lst)

#index()
print(lst.index(2))

#count()
print(lst.count(2))

lst1 = [2,312,4,5,6,7,221,2,34,653]
#sort()
lst1.sort()
print(lst1)

#reverse()
lst1.reverse()
print(lst1)

#copy()
lst2 = lst1.copy()
print(lst2)

#sum , min,max, sorted

print(sum(lst1))
print(min(lst1))
print(max(lst1))
print(sorted(lst1))

```
