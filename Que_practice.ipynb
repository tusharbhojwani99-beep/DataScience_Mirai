```python


import numpy as np

n = int(input("Number of items: "))

prices = []

for i in range(n):
    prices.append(int(input()))

prices = np.array(prices)

print("Prices =", prices)

print("Total Bill    =", np.sum(prices))
print("Average Price =", np.mean(prices))
print("Minimum Price =", np.min(prices))
print("Maximum Price =", np.max(prices))
print("Std Dev       =", np.std(prices))

print("Sorted Prices =", np.sort(prices))
print("Unique Prices =", np.unique(prices))

print("Items > 100   =", prices[np.where(prices > 100)])

```

```python
import numpy as np

n = int(input("size of rows: "))
m = int(input("size of col: "))

a = np.random.randint(-10, 10, (n, m))

sum_row = np.sum(a, axis=1)
sum_col = np.sum(a, axis=0)
ac = np.mean(a, axis=1)
ad = np.mean(a, axis=0)

z = a.copy()

for i in range(n):
    for j in range(m):
        if z[i][j] % 2 == 0:
            z[i][j] = -1

row_sorted = np.sort(a, axis=1)
col_sorted = np.sort(a, axis=0)

print("Original matrix:\n", a)
print()

for i in range(n):
    cnt_n = 0
    for j in range(m):
        if a[i][j]<0:
          cnt_n = cnt_n + 1
    print("negative columnwise : ",cnt_n)


print("Total sum of each row:", sum_row)
print("Total sum of each column:", sum_col)
print()
print("Row means:", ac)
print("Column means:", ad)
print()

print("Altered matrix:\n", z)
print()

print("Row-wise sorted matrix:\n", row_sorted)
print()

print("Column-wise sorted matrix:\n", col_sorted)
print()

```

```python
import numpy as np

n = int(input("size of rows: "))
m = int(input("size of col: "))

stock = np.random.randint(0, 10, (n, m))
print("Stock Level Matrix:\n", stock)

p = int(input("size of rows: "))
q = int(input("size of col: "))

demand = np.random.randint(-10, 10, (p, q))
print("Demand Matrix:\n", demand)

if (n, m) != (p, q):
    print("Matrices must have same dimensions!")
else:
    remain = stock - demand
    print("Remaining Stock:\n", remain)


replace = np.random.randint(0, 10, (m, n))
print("Stock Level Matrix:\n", replace)

for i in range(m):
  for j in range(n):
    if replace[i][j]<0:
      replace[i][j] = 0


print("replaced values : \n" , replace)
print()

#total shortage
sum = 0
for i in range(n):
  for j in range(m):
    if remain[i][j]<0:
      sum = sum + remain[i][j]


print("total shortage: ", sum*-1)
print()


#product with zero stock
cnt_stock = 0
for i in range(n):
  for j in range(m):
    if remain[i][j]<=0:
      cnt_stock = cnt_stock + 1

print("total product with zero stock : " , cnt_stock )
print()

#sort

row_sorted = np.sort(remain, axis=1)
col_sorted = np.sort(remain, axis=0)

print("row_sorted : ",row_sorted)
print()
print("col_sorted : ", col_sorted)


```
