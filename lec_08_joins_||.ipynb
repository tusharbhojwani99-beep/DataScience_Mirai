```python
import pandas as pd

# Option A: Adding a 5th element to match the 5 IDs
df1 = pd.DataFrame({
    'id'   : [1, 2, 3, 4, 5],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'], # Added 'amit'
    'age'  : [25, 30, 45, 67, 28]                       # Added 28
})

df2 = pd.DataFrame({
    'id'   : [3, 4, 5, 6, 7],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'], # Added 'amit'
    'age'  : [25, 32, 45, 60, 28]                       # Added 28
})

# Now the inner join will work perfectly {on common ids}

df3 = pd.merge(df1, df2, on='id', how='inner')
df3.rename(columns={'name_x': 'table_name_1', 'name_y': 'table_name_2'}, inplace=True)


print(df3)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'id'   : [1, 2, 3, 4, 5],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 30, 45, 67, 28]
})

df2 = pd.DataFrame({
    'id'   : [3, 4, 5, 6, 7],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 32, 45, 60, 28]
})


df3 = pd.merge(df1, df2, on='id', how='left')        #list down the left table data whether foun din right or not liste NaN ;
df3.rename(columns={'name_x': 'table_name_1', 'name_y': 'table_name_2'}, inplace=True)


print(df3)
```

```python
import pandas as pd

#right joins

df1 = pd.DataFrame({
    'id'   : [1, 2, 3, 4, 5],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 30, 45, 67, 28]
})

df2 = pd.DataFrame({
    'id'   : [3, 4, 5, 6, 7],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 32, 45, 60, 28]
})


df3 = pd.merge(df1, df2, on='id', how='right')
df3.rename(columns={'name_x': 'table_name_1', 'name_y': 'table_name_2'}, inplace=True)


print(df3)
```

```python
import pandas as pd
#outer joins
df1 = pd.DataFrame({
    'id'   : [1, 2, 3, 4, 5],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 30, 45, 67, 28]
})

df2 = pd.DataFrame({
    'id'   : [3, 4, 5, 6, 7],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 32, 45, 60, 28]
})


df3 = pd.merge(df1, df2, on='id', how='outer')
df3.rename(columns={'name_x': 'table_name_1', 'name_y': 'table_name_2'}, inplace=True)


print(df3)
```

```python
#cross joins

import pandas as pd

df1 = pd.DataFrame({
    'id'   : [1, 2, 3, 4, 5],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 30, 45, 67, 28]
})

df2 = pd.DataFrame({
    'id'   : [3, 4, 5, 6, 7],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 32, 45, 60, 28]
})


df3 = pd.merge(df1, df2, how='cross')
df3.rename(columns={'name_x': 'table_name_1', 'name_y': 'table_name_2'}, inplace=True)


print(df3)
```

```python
#concatination
import pandas as pd
#outer joins
df1 = pd.DataFrame({
    'id'   : [1, 2, 3, 4, 5],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 30, 45, 67, 28]
})

df2 = pd.DataFrame({
    'id'   : [3, 4, 5, 6, 7],
    'name' : ['raja', ' rani', 'sauvi', 'rhea', 'amit'],
    'age'  : [25, 32, 45, 60, 28]
})


df3 = pd.concat([df1, df2])


display(df3)

```

```python

#convert column to datetime

df = pd.DataFrame(
    {
        "date_str": ["2026-01-15", "2026-02-20", "2026-03-05"],
        "sales": [100, 150, 200],
    }
)

df["date"] = pd.to_datetime(df["date_str"])

df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day_name()

print(df)
```

```python
df = pd.DataFrame(
    {
        "date": [
            "2026-01-01",
            "2026-01-01",
            "2026-01-01",
            "2026-01-02",
            "2026-01-02",
        ],
        "city": ["Delhi", "Delhi", "Mumbai", "Delhi", "Mumbai"],
        "sales": [100, 150, 200, 120, 250],
    }
)

df3 = df.pivot_table(
    index="date", columns="city", values="sales", aggfunc="sum"
)

print(df3)
```
