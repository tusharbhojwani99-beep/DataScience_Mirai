```python
# ================================================================================
#                     DATA SCIENCE & PANDAS: JOINS COMPLETE THEORY
# ================================================================================

# [1] JOINS KYA HAI? (What is a Join?)
# --------------------------------------------------------------------------------
# Jab aapke paas do alag-alag DataFrames hote hain (manlo ek me Users ka data hai
# aur dusre me unke Purchases ka), toh unhe aapas me jodne ke liye hum Joins ka
# use karte hain.
#
# Pandas me iske liye `pd.merge()` function ka use hota hai. Yeh connection dono
# tables me kisi EK COMMON COLUMN (jise 'Key' ya 'Identifier' kehte hain,
# jaise: user_id, emp_id) ke basis par hota hai.


# [2] TYPES OF JOINS (Joins ke Prakar)
# --------------------------------------------------------------------------------

# A) INNER JOIN (how='inner') -> [Default Join
#    - Kya karta hai?: Yeh sirf wahi rows return karta hai jo DONO tables me
#      aapas me match hoti hain.
#    - Example: Agar User_ID '105' pehli table me hai par dusri table me nahi hai,
#      toh wo result me NAHI aayega.
#    - Data Science Use Case: Jab aapko ekdum clean aur complete data chahiye
#      hota hai, jahan dono sides ki information available ho.

# B) LEFT JOIN (how='left') -> [Data Science me sabse zyada use hota hai]
#    - Kya karta hai?: Yeh LEFT (pehli) table ka SAARE ke saare records uthata hai,
#      chahe wo right table me match ho ya na ho. Agar right table me match
#      nahi milta, toh wahan missing values (NaN) aa jata hai.
#    - Data Science Use Case: Jab aapka primary target master data ko bachana ho.
#      For example: Saare active users ki list chahiye, bhale hi unhone koi
#      item khareeda (Right Table) ho ya na khareeda ho.

# C) RIGHT JOIN (how='right')
#    - Kya karta hai?: Yeh bilkul Left Join ka ulta hai. Yeh RIGHT (dusri) table
#      ke saare records rakhta hai aur left table se sirf matching data lata hai.
#    - Note: Real-world data science me iska use bohot kam hota hai, kyunki log
#      tables ki position palat kar (flip karke) Left Join lagana zyada pasand karte hain.

# D) FULL OUTER JOIN (how='outer')
#    - Kya karta hai?: Yeh dono tables ka SAARE data combine kar deta hai.
#      Jahan match mila wahan data jud jayega, aur jahan match nahi mila wahan
#      dono sides me NaN bhar dega.
#    - Data Science Use Case: Exploratory Data Analysis (EDA) ke dauran, jab aapko
#      yeh dekhna ho ki dono datasets me total kitna data hai aur kahan-kahan
#      links missing hain.


# [3] DATA SCIENCE TRAPS (Joins karte waqt hone wali galtiyan)
# --------------------------------------------------------------------------------

# 1. Data Type Mismatch (Sabse badi galti):
#    Agar Table A me ID '101' ek Integer (number) hai, aur Table B me ID "101"
#    ek String (text) hai, toh join fail ho jayega. Python unhe match nahi
#    kar payega. Hamesha join se pehle data types (.astype()) check karein.

# 2. Many-to-Many Explosion (Memory Full Hone Ka Khatra
```
